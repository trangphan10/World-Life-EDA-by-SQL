```sql
select Lifeexpectancy, AdultMortality, GDP from worldlifexpectancy
where Lifeexpectancy = 0 or AdultMortality = 0 or GDP = 0
-- Một vài vị trí ở cột GDP có giá trị bằng 0, điều này là bất hợp lý. Cách xử lý ngoại lệ là thay các vị trí bằng 0 bằng giá trị trung bình của cột
```

```sql
UPDATE worldlifexpectancy
JOIN (SELECT AVG(GDP) AS avg_gdp FROM worldlifexpectancy) AS derived
SET worldlifexpectancy.GDP = derived.avg_gdp
WHERE worldlifexpectancy.GDP = 0;
```

```sql
select Lifeexpectancy, AdultMortality, GDP from worldlifexpectancy
where Lifeexpectancy = 0 or AdultMortality = 0 or GDP = 0
```

```sql
-- Basic Descriptive Stastistics
select avg(Lifeexpectancy),max(Lifeexpectancy),min(Lifeexpectancy) from worldlifexpectancy
```

```sql
SELECT (
    (SELECT MAX(Lifeexpectancy) 
     FROM (SELECT Lifeexpectancy 
           FROM worldlifexpectancy 
           ORDER BY Lifeexpectancy 
           LIMIT 719 ) AS onehalf)
    + 
    (SELECT MIN(Lifeexpectancy) 
     FROM (SELECT Lifeexpectancy 
           FROM worldlifexpectancy 
           ORDER BY Lifeexpectancy DESC 
           LIMIT 719) AS otherhalf)
) / 2 AS median;
```

Query for comparation the average lifeexpectancy between the percentage rate of Developing and Developed countries

```sql
select avg(case when Status = 'Developing' then Lifeexpectancy end) as Developing, avg(case when Status = 'Developed' then Lifeexpectancy end) as Developed
from worldlifexpectancy 
```

The average Lifeexpectancy <span style="color: #0000ff;">of</span> Developed countries <span style="color: #0000ff;">is</span> higher <span style="color: #09885a;">14</span>.<span style="color: #09885a;">7</span> units <span style="color: #0000ff;">than</span> Developing countries

```sql
-- Query for calculating the correlation between Lifeexpectancy and AdultMortality 
select (sum((Lifeexpectancy - avg_life) * (AdultMortality - avg_adult))
/sqrt(sum(pow(Lifeexpectancy - avg_life,2))*(sum(pow(AdultMortality - avg_adult,2))))) as correlation 
from worldlifexpectancy,#
(select avg(Lifeexpectancy) as avg_life,avg(AdultMortality) as avg_adult from worldlifexpectancy) as average

```

```sql
-- Set low_gdp = [2,33000], medium_gdp = [33001,50000],high_gdp = [50001,67992]
```

```sql
select avg(case when gdp between 2 and 33000 then gdp end) as low_gdp, avg(case when gdp between 33001 and 50000 then gdp end) as medium_gdp, 
avg(case when gdp between 50001 and 67992 then gdp end) as high_gdp
from worldlifexpectancy
```
